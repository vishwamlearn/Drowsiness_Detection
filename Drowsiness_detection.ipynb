```python
import cv2
import numpy as np
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
from tensorflow.keras.models import load_model
model = load_model("drowiness_detection.h5")
```

```python
cap = cv2.VideoCapture(0)
from keras.preprocessing.image import img_to_array
```

```python
while True:
    _, frame = cap.read()
    faces = face_cascade.detectMultiScale(frame, 1.1,9)
    for (x, y, w, h) in faces:
        frame=cv2.rectangle(frame, (x, y), (x+w, y+h), (0,255, 0), 1)
        roi= frame[y:y+h, x:x+w]
        roi=cv2.resize(roi, (224,224))
        roi = roi.astype('float') / 255.0
        roi = img_to_array(roi)
        roi=np.expand_dims(roi, axis=0)
        probs=model.predict(roi)
        status=probs[0].argsort(axis=-1)[0]
        if status==1:
            cv2.putText(frame, "Drowsy", (x,y-5), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 255), 1)
        else:
            cv2.putText(frame, "Non-Drowsy", (x,y-5), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 255), 1)
        
    cv2.imshow("face", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
